```python
#📦 Step 1: Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures, LabelEncoder, OneHotEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#📂 Step 2: Reading the dataset
df = pd.read_csv('./StudentPerformanceFactors.csv')

#🧹 Step 3: Explore & Clean the dataset
print("\n📌 Exploring Dataset:")
display(df)

print("\n📌 Exploring DataTypes of Features:\n")
print(df.info(),'\n')

print("\n📌 Null count:")
display(df.isnull().sum())
print("Total nulls:", df.isnull().sum().sum())

df = df.dropna(axis=0)

print("\nNulls Removed:\n Total nulls:", df.isnull().sum().sum())

print("\n📌 Exploring Unique & Frequent Values in Features:")
display(df.describe(include='all'))

#📊 Step 4: Visualize relationships
 # Visualizing the distribution of exam scores
print("\n📌 Visualizing the distribution of exam scores:")
plt.figure(figsize=(6,4))
sns.histplot(df['Exam_Score'], kde=True, bins=20)
plt.title("Distribution of Exam Scores")
plt.show()

 # Visualizing the relationship between hours studied & exam scores
print("\n📌 Visualizing the relationship between hours studied & exam scores:")
plt.figure(figsize=(6,4))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.title("Hours Studied vs Exam Score")
plt.show()

print("\nConclusion:\n Data is mostly represented in a line(linear relationship)\n")

#🎯 Step 5: Assign Target & Features
y = df[['Exam_Score']]
x = df.drop(columns=['Exam_Score'], axis=1).copy()

numerical = df.select_dtypes(include=["int64"])
categorical = df.select_dtypes(include=["object"])

#🛠️ Step 6: Feature Preparation
 # encoding categorical values
ohe = OneHotEncoder(drop='first', sparse_output=False)
encoded = ohe.fit_transform(categorical)
categorical = pd.DataFrame(encoded, columns=ohe.get_feature_names_out(categorical.columns))
 
 # regenerating index values
numerical = numerical.reset_index()
categorical = categorical.reset_index()

x = pd.concat([numerical, categorical], axis=1)

#✂️ Step 7: Splitting the dataset into training & testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)

#🤖 Step 8: Training Linear Regression Model
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)

#🔍 Step 9: Predict & Visualize Linear Regression Outputs
linear_predict = linear_model.predict(x_test)

print("\n📌 Linear Regression Results:\n")
print("MAE:", mean_absolute_error(y_test,linear_predict))
print("MSE:", mean_squared_error(y_test,linear_predict))
print("R2_Score:", r2_score(y_test, linear_predict))

 # visualize output
plt.scatter(y_test, linear_predict)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         color='red', linestyle='--', lw=2)
plt.xlabel("Actual Scores")
plt.ylabel("Predicted Scores")
plt.title("Actual vs Predicted (Linear Regression)")
plt.show()

#🤖 Step 10: Training Polynomial Regression Model
poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

poly_model = LinearRegression()
poly_model.fit(x_train_poly, y_train)

#🔍 Step 11: Predict & Visualize Polynomial Regression Outputs
poly_predict = poly_model.predict(x_test_poly)


print("\n📌 Polynomial Regression Results:\n")
print("MAE:", mean_absolute_error(y_test,poly_predict))
print("MSE:", mean_squared_error(y_test,poly_predict))
print("R2_Score:", r2_score(y_test,poly_predict))

 # visualize output
plt.scatter(y_test, poly_predict)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         color='red', linestyle='--', lw=2)
plt.xlabel("Actual Scores")
plt.ylabel("Predicted Scores")
plt.title("Actual vs Predicted (Polynomial Regression)")
plt.show()

# 📊 Step 12: Evaluate & Compare Model Performance
names = ['Linear Regression', 'Polynomial Regression']

r2_scores = [r2_score(y_test, linear_predict), r2_score(y_test, poly_predict)]
MAE_scores = [mean_absolute_error(y_test, linear_predict), mean_absolute_error(y_test, poly_predict)]
MSE_scores = [mean_squared_error(y_test, linear_predict), mean_squared_error(y_test, poly_predict)]

print("\n📌 Comparing R2-Score of Models:\n")
plt.figure(figsize=(6,4))
plt.bar(names, r2_scores, color=["#4C72B0", "#55A868"])
plt.title("R2-Score Comparison Across Models")
plt.ylabel("R2 Score")
plt.xticks(rotation=15)
plt.show()

print("\n📌 Comparing Mean Absolute Error of Models:\n")
plt.figure(figsize=(6,4))
plt.bar(names, MAE_scores, color=["#4C72B0", "#55A868"])
plt.title("Mean Absolute Error(MAE) Comparison Across Models")
plt.ylabel("MAE")
plt.xticks(rotation=15)
plt.show()

print("\n📌 Comparing Mean Squared Error of Models:\n")
plt.figure(figsize=(6,4))
plt.bar(names, MSE_scores, color=["#4C72B0", "#55A868"])
plt.title("Mean Square Error(MSE) Comparison Across Models")
plt.ylabel("MSE")
plt.xticks(rotation=15)
plt.show()


 # ✅ Final conclusion
print("\n🏆 Conclusion:")
print("Linear Regression performed slightly better due to the linear relationship of features" \
"\n as it produced the same r2_score with the least absolute error.")
```
