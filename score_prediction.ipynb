```python
#ðŸ“¦ Step 1: Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#ðŸ“‚ Step 2: Reading the dataset
df = pd.read_csv('./StudentPerformanceFactors.csv')

#ðŸ§¹ Step 3: Explore & Clean the dataset
print("\nðŸ“Œ Exploring Dataset:")
display(df)

print("\nðŸ“Œ Exploring DataTypes of Features:\n")
print(df.info(),'\n')

print("\nðŸ“Œ Exploring Unique & Frequent Values in Features:")
display(df.describe(include='all'))

print("\nðŸ“Œ Null count:")
display(df.isnull().sum())
print("Total nulls:", df.isnull().sum().sum())

df = df.dropna(axis=0)

print("\nNulls Removed:\n Total nulls:", df.isnull().sum().sum())

#ðŸ“Š Step 4: Visualize relationships
 # Visualizing the distribution of exam scores
print("\nðŸ“Œ Visualizing the distribution of exam scores:")
plt.figure(figsize=(6,4))
sns.histplot(df['Exam_Score'], kde=True, bins=20)
plt.title("Distribution of Exam Scores")
plt.show()

 # Visualizing the relationship between hours studied & exam scores
print("\nðŸ“Œ Visualizing the relationship between hours studied & exam scores:")
plt.figure(figsize=(6,4))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.title("Hours Studied vs Exam Score")
plt.show()

print("\nConclusion:\n Data is mostly represented in a line(linear relationship)\n")

#ðŸŽ¯ Step 5: Assign Target & Features
y = df[['Exam_Score']]
x = df[['Hours_Studied', 'Attendance', 'Sleep_Hours', 'Previous_Scores', 'Tutoring_Sessions', 'Physical_Activity']]

#âœ‚ï¸ Step 6: Splitting the dataset into training & testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)

#ðŸ¤– Step 7: Training Linear Regression Model
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)

#ðŸ” Step 8: Predict & Visualize Linear Regression Outputs
linear_predict = linear_model.predict(x_test)

print("\nðŸ“Œ Linear Regression Results:\n")
print("MAE:", mean_absolute_error(y_test,linear_predict))
print("MSE:", mean_squared_error(y_test,linear_predict))
 # storing r2 score for comparison
r2_scores = [r2_score(y_test, linear_predict)]
print("R2_Score:", r2_scores[0])

 # visualize output
plt.scatter(y_test, linear_predict)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         color='red', linestyle='--', lw=2)
plt.xlabel("Actual Scores")
plt.ylabel("Predicted Scores")
plt.title("Actual vs Predicted (Linear Regression)")
plt.show()

 # 2nd visualization of output
plt.plot(range(len(y_test)), y_test.values, label="Actual")
plt.plot(range(len(linear_predict)), linear_predict, label="Predicted")
plt.legend()
plt.title("Actual vs Predicted (Linear Regression)")
plt.show()

#ðŸ¤– Step 9: Training Polynomial Regression Model
poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

poly_model = LinearRegression()
poly_model.fit(x_train_poly, y_train)

#ðŸ” Step 10: Predict & Visualize Polynomial Regression Outputs
poly_predict = poly_model.predict(x_test_poly)


print("\nðŸ“Œ Polynomial Regression Results:\n")
print("MAE:", mean_absolute_error(y_test,poly_predict))
print("MSE:", mean_squared_error(y_test,poly_predict))
 # storing r2 score for comparison
r2_scores.append(r2_score(y_test,poly_predict))
print("R2_Score:", r2_scores[1])

 # visualize output
plt.scatter(y_test, poly_predict)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         color='red', linestyle='--', lw=2)
plt.xlabel("Actual Scores")
plt.ylabel("Predicted Scores")
plt.title("Actual vs Predicted (Polynomial Regression)")
plt.show()

 # 2nd visualization of output
plt.plot(range(len(y_test)), y_test.values, label="Actual")
plt.plot(range(len(poly_predict)), poly_predict, label="Predicted")
plt.legend()
plt.title("Actual vs Predicted (Polynomial Regression)")
plt.show()

# ðŸ“Š Step 15: Evaluate & Compare Model Performance
print("\nðŸ“Œ Comparing R2-Score of Models:\n")

 # Compare F1-scores from the 3 approaches
names = ['Linear Regression', 'Polynomial Regression']

plt.figure(figsize=(6,4))
plt.bar(names, r2_scores, color=["#4C72B0", "#55A868"])
plt.title("R2-Score Comparison Across Models")
plt.ylabel("R2 Score")
plt.xticks(rotation=15)
plt.show()

 # âœ… Final conclusion
print("\nðŸ† Conclusion:")
print("Linear Regression performed slightly better due to the linear relationship of features.")
```
